```python
from GoogleNews import GoogleNews


words_offensive = ['gay','lesbian','asisan','white','hispanic']

main_textlist = []
def get_news(word):
    googlenews = GoogleNews()
    googlenews.get_texts()
    googlenews = GoogleNews(lang='en')
    googlenews = GoogleNews(lang='en', region='US')
    googlenews = GoogleNews(period='7d')
    googlenews = GoogleNews(encode='utf-8')
    googlenews.set_lang('en')
    googlenews.set_period('7d')
    googlenews.set_time_range('02/01/2020','02/28/2020')
    googlenews.set_encode('utf-8')
    googlenews.get_news(word)
    googlenews.search(word)
    result = googlenews.page_at(2)
    googlenews.results()
    main_textlist.append(googlenews.get_texts())
    print(len(main_textlist))
    googlenews.clear()
    
for i in words_offensive:
    get_news(i)
```
