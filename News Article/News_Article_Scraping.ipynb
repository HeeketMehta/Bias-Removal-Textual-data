## Load balanced_data.csv (training data) to train logistic model

```python
import pandas as pd
import numpy as np
def converter(instr):
    return np.fromstring(instr[1:-1],sep=' ')
df1=pd.read_csv('/content/gdrive/MyDrive/balanced_train_bert_encoding.csv',converters={'vectors':converter})
```

```python
from sklearn.linear_model import LogisticRegression
model_identity_hate_lr = None
model_toxic_lr = None

import copy
def baseline_ml(target):
  X = df1['vectors']
  #print(type(X[1]))
  y = df1[target]
  #print(y)
  print("Logistic Regression for -",target)
  lst = []
  for i in X:
    lst.append(i)
  #print(type(lst[2]))

  from sklearn.model_selection import train_test_split
  x_train, x_test, y_train, y_test = train_test_split(lst, list(y), test_size=0.25, random_state=0)

  model = LogisticRegression(max_iter = 10000)
  model.fit(x_train, y_train)
  #Prediction using Logistic Reg
  if target == 'toxic':
    model_toxic_lr = copy.deepcopy(model)
    
  elif target == 'identity_hate':
    model_identity_hate_lr = copy.deepcopy(model)

  predictions = model.predict(x_test)
  
  import matplotlib.pyplot as plt
  import seaborn as sns
  from sklearn import metrics
  cm = metrics.confusion_matrix(y_test, predictions)
  print("Confusion Mtarix:")
  print(cm)
  precision = metrics.precision_score(y_test, predictions)
  recall = metrics.recall_score(y_test, predictions)
  F1 = 2 * (precision * recall) / (precision + recall)

  print("Accuracy of {} -- {}".format(target, metrics.accuracy_score(y_test, predictions)))
  print("Precision:",precision)
  print("Recall:",recall)
  print("F1 Score - ",F1)

  print("===============")
  print()

  return model
```

```python
model_toxic_lr = baseline_ml('toxic')
model_identity_hate_lr = baseline_ml('identity_hate')
```

## Scrape and clean news articles

```python
# to tokenize paragraphs
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize
```

```python
response = get('https://www.foxnews.com/opinion/being-black-and-being-gay-are-not-the-same-thing')
extractor = Goose()
article = extractor.extract(raw_html=response.content)
text = article.cleaned_text
text
```

```python
text = text.replace("\\","")
text = text.replace("\n", "")
text = text.replace("\.", ". ")
text
```

```python
# split data into sentences
#sent_text = sent_tokenize(text)
sent_text = text.split(".")
sent_text
```

```python
# make a dataframe of sentences
df2 = pd.DataFrame()
df2['Sentences'] = sent_text
df2
```

## Logistic Regression Prediction - BERT

```python
# predict toxicity on the scraped data using the trained logistic model
df_temp = pd.DataFrame()
l_s = []
l_pred = []
for i in range(len(df2['Sentences'])):
  embs = sbert_model.encode(df2['Sentences'][i])
  prediction = model_toxic_lr.predict(embs.reshape(1,-1))  
  l_s.append(df2['Sentences'][i])
  l_pred.append(prediction[0])
```

```python
df_temp['sent'] = l_s
df_temp['preds'] = l_pred
```

## Custom tagging on Article
Perform custom tagging on the sentences 


```python

lst_embeddings = []
for i in range(len(df2['Sentences'])):
  sentences_embeddings = sbert_model.encode(df2['Sentences'][i])
  lst_embeddings.append(sentences_embeddings)
  
```

```python
df2['vectors'] = lst_embeddings
```

```python
# define custom labels
lst_biased_words = ['black','bi-sexual','gay','lesbian','homoesexual','asian']
offensive_word_embeddings = sbert_model.encode(lst_biased_words)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
for i in lst_biased_words:
  df2[i] = ''
```

```python
from sklearn.metrics.pairwise import cosine_similarity
for each_vector_idx in range(len(df2['vectors'])):
  for each_word_idx in range(len(offensive_word_embeddings)):
    cos_sim = cosine_similarity(offensive_word_embeddings[each_word_idx].reshape(1,-1), np.array(df2['vectors'][each_vector_idx]).reshape(1,-1))
    if cos_sim>=0.45:
      df2[lst_biased_words[each_word_idx]][each_vector_idx] = 1
    else:
      df2[lst_biased_words[each_word_idx]][each_vector_idx] = 0
```
