```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')

# We tokenize the sentences that are in the form of paragraphs.
# We take roughly 1000 sentences from the 450,000 datapoints, for variety of dataset, to introduce both bias and unbiased sentences.
newdf = df[:1000]
newlist = []
for i in range(len(newdf['comment_text'])):
  sent_text = sent_tokenize(newdf['comment_text'][i])
  newlist.append(sent_text[0])
```

```python
# biased_data.csv  - google scraped
df_bias = pd.read_csv(r'/content/gdrive/MyDrive/biased data.csv')
```

## We combine the two dataframes - biased.csv and shanay_train_vec

```python
# 0 is the column name for sentences
lst_bias = []
for i in df_bias['0']:
  lst_bias.append(i)
```

```python
# combining sentences from google and shanay_train_vec
lst_combined = []
for i in newlist:
  lst_combined.append(i)
for i in lst_bias:
  lst_combined.append(i)
```

```python
len(lst_combined)
```

```python
# new dataframe df2 having combination of both datasets
df2 = pd.DataFrame()
df2['sentences'] = lst_combined
df2
```

## Find embeddings

```python

!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
# encode using BERT
lst_embeddings = []
for idx in range(len(df2['sentences'])):
  if idx % 500 == 0:
    print(idx)
  sentences_embeddings = sbert_model.encode(df2['sentences'][idx])
  lst_embeddings.append(sentences_embeddings)
```

```python
df2['vectors'] = lst_embeddings
#df_combined.to_csv('/content/gdrive/MyDrive/jigsaw-unintended-bias-in-toxicity-classification/try.csv')
```

```python
df2
```

## Custom Tagging

```python
import numpy as np
```

```python
lst_words = ['gay', 'black', 'lesbian', 'chinese', 'indian']
offensive_embeddings = sbert_model.encode(lst_words)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
for i in lst_words:
  df2[i] = ''
```

```python
df2
```

```python
from sklearn.metrics.pairwise import cosine_similarity
for i in lst_words:
  df2[i] = ''
for each_vector_idx in range(len(df2['vectors'])):
  for each_word_idx in range(len(offensive_embeddings)):
    cos_sim = cosine_similarity(offensive_embeddings[each_word_idx].reshape(1,-1), np.array(df2['vectors'][each_vector_idx]).reshape(1,-1))
    if cos_sim>=0.35:
      df2[lst_words[each_word_idx]][each_vector_idx] = 1
    else:
      df2[lst_words[each_word_idx]][each_vector_idx] = 0
```

```python
df2
```

```python
# This file/dataframe gives us the tagging of 2818 sentences, and their custom tagging based on 5 classes, where potential bias may be introduced.
df2 = df2.sample(frac=1)
df2.to_csv("/content/gdrive/MyDrive/google_shanay_train_vec.csv", index=False)
```
